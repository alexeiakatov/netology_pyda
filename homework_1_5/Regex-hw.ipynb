## Задание 1

Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером. Если да, то функция должна возвращать отдельно номер и регион.

```python
import re

valid_car_numbers = ['е893хт59', 'у987вн177', 'е177кв26', 'о007оо77', 'а145нв147', 'т533ре123']
not_valid_car_numbers = ['те335хн123', '335хн123', 'т3кв123', 'т39кв123', 'т399квт123', 'т399123', 'т399кв', 'т399кв1', 'т399кв3349', '_т399кв99а']

car_number_validation_pattern = '^([авекмнорстху]{1}\d{3}[авекмнорстху]{2})(\d{2,3})$'


def car_number_validator(car_number):
    car_number = car_number.lower().strip()
    result = re.search(car_number_validation_pattern, car_number)

    if result is not None:
        print(f'Валидный номер: {result.group(0)}')
        print(f'\tНомер: {result.group(1)}')
        print(f'\tРегион: {result.group(2)}')
        print()
        return

    print(f'Не валидный номер: {car_number}')


for car_number in valid_car_numbers:
    car_number_validator(car_number)

print('*********')

for car_number in not_valid_car_numbers:
    car_number_validator(car_number)
```

## Задание 3
Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки.

```python
import re

arr = [
    'Маша ела кашу',
    'Маша Маша Маша Маша ела ела ела ела кашу кашу кашу кашу',
    'Маша ела кашу кашу',
    'Маша Маша Маша Маша Маша Маша Маша Маша Маша ела кашу',
    'Маша ела ела кашу'
    ]

pattern = r'(([а-яА-я]+)(\s+\2)+)+'

for phrase in arr:
    print('******************')
    print(f'фраза: \r\n\t{phrase}')
    result = re.sub(pattern, r'\2', phrase)
    print(f'результат: \r\n\t{result}')
```

## Задание 4
Напишите функцию, которая будет проверять номер сотового телефона на валидность, если он валиден, то переводить его в формат:   
+7(xxx)-xxx-xx-xx  
Постарайтесь предусмотреть как можно больше адекватных форматов изначального ввода номера.
Примеры правильных и неправильных форматов:
- +7 955 555-55-55;
- 955555555;
- 8(955)555-55-55;
- +7 955 555 55 55;
- 7(955) 555-55-55;
- +7 955+555+55+55.

```python
import re


phone_numbers_with_prefix = [
    '+7 955 555-55-51',
    '8(955)555-55-51',
    '+7 955 555 55 51',
    '7(955) 555-55-51',
    '+7 955+555+55+51',
    '+7~9z5z5 5t5 5n5n5d5d1',
    '8~9~5~5~5~5~5~5~5~5~1'
]

phone_numbers_without_prefix = [
    '9555 555551',
    '9z5z5 5t5 5n5n5d5d1',
]


def search_with_prefix(phone_numbers):
    check_pattern = r'^(8|\+7|7)(([^\d]{0,2})\d{1}){10}$'

    print('Форматирование номеров с префиксами: +7, 7, 8')
    print()

    for phone_number in phone_numbers:
        match = re.match(check_pattern, phone_number)

        if match is not None:
            print(f'Номер: \r\n\t{phone_number}')
            digits = re.sub(r'[^\d]', '', phone_number)
            print('Формат: ')
            print(f'\t+7({digits[1:4]})-{digits[4:7]}-{digits[7:9]}-{digits[9:]}')
            print('****************************')


def search_without_prefix(phone_numbers):
    check_pattern = '^([^\d]{0,2}\d{1}){10}$'

    print('Форматирование номеров без префиксов: +7, 7, 8')
    print()

    for phone_number in phone_numbers:
        match = re.match(check_pattern, phone_number)

        if match is not None:
            print(f'Номер: \r\n\t{phone_number}')
            digits = re.sub(r'[^\d]', '', phone_number)

            print('Формат: ')
            print(f'\t+7({digits[0:3]})-{digits[3:6]}-{digits[6:8]}-{digits[8:]}')
            print('****************************')



search_with_prefix(phone_numbers_with_prefix)
print()
print()
search_without_prefix(phone_numbers_without_prefix)
```
